Trong source page có một script tag với id `__NEXT_DATA__` chứa thông tin query của page tiếp theo, dựa vào đó ta có đoán các query param:

```js
            "o": 40,
            "page": 3,
            "st": "u%2Ch",
            "limit": "20",
            "w": "1"

```


API url có dạng như sau: [https://gateway.chotot.com/v1/public/ad-listing?region_v2=13000&cg=1000&limit=1&o=0&st=s,k&page=0](https://gateway.chotot.com/v1/public/ad-listing?region_v2=13000&cg=1000&limit=1&o=0&st=s,k&page=0)

Trong đó:
+ `https://gateway.chotot.com/v1/public/ad-listing?` : API Endpoint
+ Các query param:
    + **region_v2**: mã TP/tỉnh VD: 13000 là HCM, 12000 là HN
    + **cg**: mã category VD 1010 tương ứng chung cư
    + **limit**: giới hạn số record (pagination)
    + **o**: bắt đầu từ record số mấy (pagination)
    + **page**: trang bao nhiêu (pagination)
    + **st**: phân loại (s,k: mua bán, u,h: cho thuê)

### 2.2. CONSTANT

````python
BASE_URL = 'https://gateway.chotot.com/v1/public/ad-listing'

LIMIT = 100
# Vị trí: HCM
REGION_V2 = 13000
# Category: Bất động sản
CG = 1000
# Type: Mua bán
ST = "s,k"

# MongoDB constants
ATLAS_URL = "mongodb+srv://conghuynt1999:foWcElkNDeLlDNIX@cluster1.oq1l7.mongodb.net/?retryWrites=false&replicaSet=atlas-5tjaoq-shard-0"
DATABASE_NAME = "chotot"
COLLECTION_NAME = "processed_data"
````

### 2.3. Lấy danh sách bất động sản

#### Hàm crawl data (240000 dòng)

````python
def fetch_data(city_code=REGION_V2):
    """Fetches data from Cho Tot listings for the given city code.

    Args:
        city_code (int, optional): The city code to fetch data for. Defaults to 13000 (Ho Chi Minh City).

    Returns:
        list: A list of dictionaries containing the fetched data.
    """

    error_limit = 6
    user_agents = [
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36',
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36',
        'Mozilla/5.0 (iPhone; CPU iPhone OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148',
        'Mozilla/5.0 (Linux; Android 11; SM-G960U) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.72 Mobile Safari/537.36'
    ]

    data = []
    error = 0

    while True:
        page = 0
        o = -LIMIT
        print('Scanning...')

        while True:
            try:
                page += 1
                o += LIMIT
                url = f"{BASE_URL}region_v2={city_code}&cg=1000&o={o}&page={page}&st={ST}&limit={LIMIT}&key_param_included=true"
                headers = {'User-Agent': random.choice(user_agents)}
                response = requests.get(url, headers=headers)
                response.raise_for_status()  # Check for request errors

                if not response.json()['ads'] or len(data) >= 220000:
                    break

                data.extend(response.json()['ads'])
                print(f'Number of items: {page * LIMIT} (Total: {len(data)})')
            except requests.exceptions.RequestException as e:
                print(f"Error: {e}")

            time.sleep(np.random.choice([x / 100 for x in range(3, 12)]))

        if page == 1:
            error += 1

        if error > error_limit:
            break

    print(f'\nFinish {len(data)} items\n')
    return data
````

#### Hàm lưu data vào file csv

````python
def save_data_to_csv(data, csv_file_path):
    """Saves the collected data to a CSV file.

    Args:
        data (list): A list of dictionaries containing the data to save.
        csv_file_path (str, optional): The path to the CSV file. Defaults to 'ads.csv'.
    """

    with open(csv_file_path, 'a', newline='', encoding='utf-8') as csvfile:
        fieldnames = set()
        for row in data:
            fieldnames.update(row.keys())

        # Check for expected columns
        expected_columns = data[0].keys() if data else []
        for col in expected_columns:
            if col not in fieldnames:
                fieldnames.add(col)

        writer = csv.DictWriter(csvfile, fieldnames=list(fieldnames))
        if not os.path.exists(csv_file_path) or os.stat(csv_file_path).st_size == 0:
            writer.writeheader()
        writer.writerows(data)
    print(f"Data saved to {csv_file_path}")
````

#### Tiến hành crawl data và lưu vào file csv

````python
save_data_to_csv(fetch_data(), './data/raw_data.csv')
````

#### Load data và những cột cần thiết

````python
file_path = './data/raw_data.csv'

selected_columns = ['list_id', 'price', 'price_million_per_m2', 'size', 'width', 'length', 'area','subject', 'body', 'category_name', 'latitude', 'list_time','longitude', 'toilets', 'rooms', 'floors', 'ward_name', 'area_name', 'street_name', 'street_number']

df = pd.read_csv(file_path, usecols=selected_columns)
df = df[selected_columns]
len(set(df['list_id']))
````

#### Xử lý dữ liệu ở cột `subject` và `body`

````python
icon_pattern = re.compile(r'[^\w\s!,.;?]+|\s\s+|\t+')
df['subject'] = df['subject'].apply(lambda x: icon_pattern.sub('', x.replace('\n', ' ')))
df['body'] = df['body'].apply(lambda x: icon_pattern.sub('', x.replace('\n', ' ')))
df
````

### 2.5. Lưu data vào MongoDB

````python
db = pymongo.MongoClient(ATLAS_URL)[DATABASE_NAME]
collection = db.get_collection(COLLECTION_NAME)

batch_size = 2000

# Insert data in batches
for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i+batch_size].to_dict('records')
    collection.insert_many(batch)
````
